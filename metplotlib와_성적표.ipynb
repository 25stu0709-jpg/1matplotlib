```python
import matplotlib.pyplot as plt

# 학생 데이터 (예시)
data = [
    ["name", "korean", "math", "science", "total", "average"],
    ["park", 87, 98, 87, 272, 90.67],
    ["moon", 85, 90, 90, 265, 88.33],
    ["kim", 87, 98, 67, 252, 84.00],
    ["lee", 98, 87, 98, 283, 94.33],
    ["shin", 87, 100, 98, 285, 95.00],
    ["average", 88.8, 94.6, 88.0, "", ""]
]

fig, ax = plt.subplots()

# 축 숨기기
ax.axis('off')

# 표 그리기
table = ax.table(cellText=data, loc='center', cellLoc='center')

table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2, 1.2)

plt.title("student score")
plt.show()
```
